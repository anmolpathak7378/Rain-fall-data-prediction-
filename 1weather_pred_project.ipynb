```python
#here is a clear visulization of rain-fall data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
from sklearn.linear_model import LinearRegression

```

```python
#now will import the data 

data = pd.read_csv("austin_weather.csv")
data.head()
```

```python
#now remove unnecessary columns

data = data.drop(["Date", "Events"], axis=1)

#remove null values

```

```python
data = data.replace("T", 0)
```

```python
data = data.replace("-", 0)
```

```python
data
```

```python
#convert this data 

data.to_csv("weather_final.csv")
```

```python
#import this final data

data = pd.read_csv("weather_final.csv")
data
```

```python
#make labels
X = data.drop(["PrecipitationSumInches"], axis= 1)
```

```python
Y = data["PrecipitationSumInches"]
```

```python
#reshape the value of Y in 2d vector

Y = Y.values.reshape(-1, 1)
Y
```

```python
#now we will take any random day for prediction
day_index = 798
days = [i for i in range(Y.size)]
days
```

```python
#initialize linier reg

clf = LinearRegression()
clf.fit(X,  Y)
```

```python
#plot the graphs

plt.scatter(days, Y, color = "green")
plt.scatter(days[day_index], Y[day_index], color = "red")
plt.scatter(days[200], Y[200], color = "blue")

plt.title("the precipitation level")
plt.xlabel("days")
plt.ylabel("PrecipitationSumInches")
plt.show()
```

```python
x_vis = X.filter(["DewPointHighF", "TempAvgF", "VisibilityAvgMiles", "WindAvgMPH", "SeaLevelPressureAvgInches", "HumidityHighPercent"])

print("adding subplot - the precipitation vs attributes")

plt.figure(figsize=(20,15))  # Set the overall figure size
plt.yticks(rotation=0)  # or try 45/90 to suit your style

for i in range(x_vis.columns.size):
    plt.subplot(3, 3, i + 1) 
    plt.scatter(days, x_vis.iloc[:, i], color="yellow")
    plt.scatter(days[day_index], x_vis.iloc[day_index, i], color="red")
    plt.title(x_vis.columns[i])
    plt.xlabel("Days")
    plt.ylabel("Precipitation in Inches")



plt.tight_layout()
plt.show()
```
